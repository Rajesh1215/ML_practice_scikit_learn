```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Generate synthetic data
np.random.seed(42)

# Class 0 (red circles)
X_class0 = np.random.randn(20, 2) + [1, 1]
y_class0 = np.zeros(20)

# Class 1 (blue squares)
X_class1 = np.random.randn(20, 2) + [3, 3]
y_class1 = np.ones(20)

# Combine the data for both classes
X = np.vstack((X_class0, X_class1))
y = np.concatenate((y_class0, y_class1))

# Create an SVM classifier
svm_classifier = SVC(kernel='rbf')
svm_classifier.fit(X, y)

# Generate new data points for visualization
X_new = np.array([[0, 0], [4, 4], [2, 2]])

# Predict the class labels for new data points
y_pred = svm_classifier.predict(X_new)

# Plot the data points and the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
plt.scatter(X_new[:, 0], X_new[:, 1], c=y_pred, marker='x', s=100, cmap=plt.cm.Spectral)
plt.xlabel('Feature X1')
plt.ylabel('Feature X2')
plt.title('Support Vector Machines (SVM)')
plt.show()
print(X)
X[:, 0]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Generate synthetic data
np.random.seed(42)

# Class 0 (red circles)
X_class0 = np.random.randn(20, 2) + [1, 1]
y_class0 = np.zeros(20)

# Class 1 (blue squares)
X_class1 = np.random.randn(20, 2) + [3, 3]
y_class1 = np.ones(20)

# Combine the data for both classes
X = np.vstack((X_class0, X_class1))
y = np.concatenate((y_class0, y_class1))

# Create SVM classifiers with different values of C
svm_classifier_strict = SVC(kernel='linear', C=100)  # Larger C enforces a stricter margin (hard margin)
svm_classifier_soft = SVC(kernel='linear', C=0.1)    # Smaller C allows a softer margin (soft margin)

# Fit the classifiers to the data
svm_classifier_strict.fit(X, y)
svm_classifier_soft.fit(X, y)

# Generate new data points for visualization
X_new = np.array([[0, 0], [4, 4], [2, 2]])

# Predict the class labels for new data points
y_pred_strict = svm_classifier_strict.predict(X_new)
y_pred_soft = svm_classifier_soft.predict(X_new)

# Create a meshgrid for decision boundary visualization
xx, yy = np.meshgrid(np.linspace(-2, 5, 100), np.linspace(-2, 5, 100))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Predict the class labels for the grid points
grid_predictions_strict = svm_classifier_strict.predict(grid_points)
grid_predictions_soft = svm_classifier_soft.predict(grid_points)

# Reshape the predictions to match the grid shape
grid_predictions_strict = grid_predictions_strict.reshape(xx.shape)
grid_predictions_soft = grid_predictions_soft.reshape(xx.shape)

# Plot the data points, decision boundary, and margins for strict and soft SVM classifiers
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.contourf(xx, yy, grid_predictions_strict, cmap=plt.cm.Spectral, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
plt.scatter(X_new[:, 0], X_new[:, 1], c=y_pred_strict, marker='x', s=100, cmap=plt.cm.Spectral)
plt.title('Strict Margin SVM (C=100)')
plt.xlabel('Feature X1')
plt.ylabel('Feature X2')

plt.subplot(1, 2, 2)
plt.contourf(xx, yy, grid_predictions_soft, cmap=plt.cm.Spectral, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
plt.scatter(X_new[:, 0], X_new[:, 1], c=y_pred_soft, marker='x', s=100, cmap=plt.cm.Spectral)
plt.title('Soft Margin SVM (C=0.1)')
plt.xlabel('Feature X1')
plt.ylabel('Feature X2')

plt.tight_layout()
plt.show()
```
